```python
import os, zipfile, glob
from collections import Counter
from google.colab import drive

os.makedirs('/content/drive/MyDrive/PRACTICUM DATASET/data', exist_ok=True)

# >>> EDIT THIS PATH to your folder in Drive <
ZIP_FOLDER = '/content/drive/MyDrive/PRACTICUM DATASET'

for zp in glob.glob(f'{ZIP_FOLDER}/*.zip'):
    with zipfile.ZipFile(zp) as z:
        z.extractall('/content/drive/MyDrive/PRACTICUM DATASET/data')
    print('Unzipped:', os.path.basename(zp))

exts = Counter(os.path.splitext(f)[1].lower() for f in glob.glob('data/**/*', recursive=True) if os.path.isfile(f))
print('File types:', dict(exts))
```
